```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.patches import Rectangle

dataset_path = os.getcwd() + '/Basketball/'
FACTOR = 0.7
IMAGE_WIDTH = int(576 * FACTOR)
IMAGE_HEIGHT = int(432 * FACTOR)

def read_images():
    images_path = dataset_path + 'img/'
    image_names = os.listdir(images_path)
    image_names = sorted(image_names)
    images = []
    for _, image_name in enumerate(image_names):
        image = cv2.imread(images_path + image_name)
        image = cv2.resize(image, (IMAGE_WIDTH, IMAGE_HEIGHT))
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        images.append(image)
        
    gt_path = dataset_path + 'groundtruth_rect.txt'
    f = open(gt_path, "r")
    
    labels = []
    for line in f:
        strip_string = line.rstrip()
        array = strip_string.split(',')
        array = [ int(x) for x in array ]
        array = [ x * FACTOR for x in array ]

        labels.append(array)
    
    return np.array(images), np.array(labels)
```

```python
images, labels = read_images()
```

```python
import keras
from keras.models import Model
from keras.layers import Input, Conv2D, TimeDistributed, Flatten, LSTM, Dense, MaxPooling2D
```

```python
image_input = Input(shape = (None, IMAGE_HEIGHT, IMAGE_WIDTH, 3))

x = TimeDistributed(Conv2D(filters= 24, activation='relu', kernel_size=(3, 3))) (image_input)
x = TimeDistributed(MaxPooling2D()) (x)
x = TimeDistributed(Conv2D(filters= 24, activation='relu', kernel_size=(3, 3))) (x)
x = TimeDistributed(MaxPooling2D()) (x)
x = TimeDistributed(Conv2D(filters= 24, activation='relu', kernel_size=(3, 3))) (x)
x = TimeDistributed(MaxPooling2D()) (x)
x = TimeDistributed(Conv2D(filters= 24, activation='relu', kernel_size=(3, 3))) (x)
x = TimeDistributed(MaxPooling2D()) (x)
x = TimeDistributed(Conv2D(filters= 24, activation='relu', kernel_size=(3, 3))) (x)
x = TimeDistributed(MaxPooling2D()) (x)

x = TimeDistributed(Flatten()) (x)
x = LSTM(10, return_sequences = True) (x)
x = TimeDistributed(Dense(4)) (x)

model = Model(image_input, x)
```

```python
from keras.optimizers import Adam
model.compile(optimizer = Adam(0.0001), loss = 'mse')
# model.summary()
```
